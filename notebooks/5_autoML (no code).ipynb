# Start your first autoML experiment 
* Follow instructions in notebook1 to finish upload the diabetes_modified dataset to your created database on Databricks
* On top left (the icon below databricks logo), choose "machine learning" persona. Note, if you are using **Databricks community edition**, you **may not** be able to run autoML
* Click create->autoML experiment
* Follow the steps, for example, choose a cluster to attach to, choose your data set, classification, etc.
* Start running autoML experiments and review results and also the automatically generated EDA notebook
